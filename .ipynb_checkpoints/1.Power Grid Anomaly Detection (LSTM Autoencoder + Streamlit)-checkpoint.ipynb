| Label | Category | Description |
|-------|----------|-------------|
| 0 | Normal | All values within normal operating range |
| 1 | Voltage Anomaly | Abnormal voltage (sag/surge) |
| 2 | Frequency Anomaly | Frequency deviates from nominal 50Hz |
| 3 | Power Factor Anomaly | Power factor abnormally low (< 0.7) |
| 4 | Combined Anomaly | Two or more anomalies present simultaneously |

```python
from tensorflow.keras.preprocessing import timeseries_dataset_from_array
```

```python
y = y.reshape(-1,1)
```

```python
xy_combined = np.hstack((x_scaled, y))
```

```python
xy_combined
```

```python
tf_seq = timeseries_dataset_from_array(
    xy_combined,
    targets=None,
    sequence_length=12,
    sequence_stride=1,
    batch_size=32
)
```

```python
sequences = []
for batch in tf_seq:
    sequences.append(batch.numpy())
```

```python
all_windows = np.concatenate(sequences, axis=0) 
```

```python
all_windows.shape
```

```python
x_windows = all_windows[:, :, :-1] 
y_windows = all_windows[:, :, -1]
```

```python
x_windows.shape
```

```python
y_windows
```

```python
pure_normal_idx   = np.where((y_windows == 0).all(axis=1))[0]
mixed_or_anom_idx = np.where((y_windows != 0).any(axis=1))[0]

x_pure_normal = x_windows[pure_normal_idx]
y_pure_normal = y_windows[pure_normal_idx] 

x_anomaly   = x_windows[mixed_or_anom_idx]
y_anomaly   = y_windows[mixed_or_anom_idx]
```

```python
y_anomaly = np.max(y_anomaly, axis=1) 
y_anomaly = y_anomaly - 1 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train_anomaly, x_temp_anomaly, y_train_anomaly, y_temp_anomaly = train_test_split(
    x_anomaly, y_anomaly, test_size=0.4, random_state=42, stratify=y_anomaly
)
x_val_anomaly, x_test_anomaly, y_val_anomaly, y_test_anomaly = train_test_split(
    x_temp_anomaly, y_temp_anomaly, test_size=0.5, random_state=42, stratify=y_temp_anomaly
)
```

```python
x_train_normal, x_temp_normal, y_train_normal, y_temp_normal = train_test_split(
    x_pure_normal, y_pure_normal, test_size=0.4, random_state=42
)
x_val_normal, x_test_normal, y_val_normal, y_test_normal = train_test_split(
    x_temp_normal, y_temp_normal, test_size=0.5, random_state=42
)
```

```python
def to_tensor(x, y):
    return torch.FloatTensor(x), torch.LongTensor(y)
```

```python
# Autoencoder uses ONLY normal data for training
x_train_norm_t, y_train_norm_t = to_tensor(x_train_normal, y_train_normal)
x_val_norm_t, y_val_norm_t = to_tensor(x_val_normal, y_val_normal)
x_test_norm_t, y_test_norm_t = to_tensor(x_test_normal, y_test_normal)

# Classifier uses ONLY anomaly data
x_train_anom_t, y_train_anom_t = to_tensor(x_train_anomaly, y_train_anomaly)
x_val_anom_t, y_val_anom_t = to_tensor(x_val_anomaly, y_val_anomaly)
x_test_anom_t, y_test_anom_t = to_tensor(x_test_anomaly, y_test_anomaly)
```

```python
x_train_normal.shape
```

```python
from torch.utils.data import TensorDataset, DataLoader
```

```python
batch_size=64
```

```python
train_normal_loader = DataLoader(TensorDataset(x_train_norm_t, x_train_norm_t), batch_size=batch_size, shuffle=True)
val_normal_loader = DataLoader(TensorDataset(x_val_norm_t, x_val_norm_t), batch_size=batch_size, shuffle=False)
test_autoencoder_loader = DataLoader(TensorDataset(x_test_norm_t, x_test_norm_t), batch_size=batch_size, shuffle=False)
```

```python
train_anomaly_loader = DataLoader(TensorDataset(x_train_anom_t, y_train_anom_t), batch_size=batch_size, shuffle=True)
val_anomaly_loader = DataLoader(TensorDataset(x_val_anom_t, y_val_anom_t), batch_size=batch_size, shuffle=False)
```

```python
test_classifier_loader = DataLoader(TensorDataset(x_test_anom_t, y_test_anom_t), batch_size=batch_size, shuffle=False)
```

```python
class autoencoder(nn.Module):

    def __init__(self, input_size, hidden_size, num_layers):
        super().__init__()
        self.input_size=input_size
        self.hidden_size=hidden_size
        self.num_layers=num_layers
        self.encoder = nn.LSTM(self.input_size,self.hidden_size,self.num_layers,batch_first=True)
        self.decoder = nn.LSTM(input_size=self.hidden_size, hidden_size=self.hidden_size, num_layers=self.num_layers, batch_first=True)
        self.output_layer = nn.Sequential(
            nn.Linear(in_features=self.hidden_size, out_features=32),
            nn.ReLU(),
            nn.Linear(32, self.input_size)
        )
    def forward(self,seq):
        _,(hidden,cell) = self.encoder(seq)
        bottleneck = hidden[-1]
        bottleneck = bottleneck.unsqueeze(1)
        bottleneck = bottleneck.repeat(1,12,1)
        output,(hid,cell2) = self.decoder(bottleneck)
        out = self.output_layer(output)
        return out
```

```python
auto = autoencoder(5,64,1)
auto.cuda()
```

```python
epochs = 1000
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(auto.parameters(), lr=1e-4)
```

```python
train_loss=[]
val_loss=[]

for i in range(epochs):
    n=0
    auto.train()
    train_batch_loss=[]
    for batch,(x,y) in enumerate(train_normal_loader):
        n+=1
        x = x.cuda()

        optimizer.zero_grad()
        reconstruct = auto(x)
        losses = criterion(x, reconstruct)
        losses.backward()
        optimizer.step()

        train_batch_loss.append(losses.item())
    train_loss.append(sum(train_batch_loss)/n)

    auto.eval()
    n=0
    val_batch_loss=[]
    with torch.no_grad():
        for batch,(x,y) in enumerate(val_normal_loader):
            n+=1
            x = x.cuda()
            reconstruct = auto(x)
            losses = criterion(x,reconstruct)

            val_batch_loss.append(losses.item())
    val_loss.append(sum(val_batch_loss)/n)

    print(f'epoch {i+1}, train_loss: {train_loss[-1]}, val_loss: {val_loss[-1]}')
```

```python
errors_normal = np.array(individual_normal_error)
errors_anom   = np.array(individual_anomaly_error)
```

```python
errors = np.concatenate([errors_normal, errors_anom])
y_true = np.concatenate([
    np.zeros_like(errors_normal, dtype=int),  
    np.ones_like(errors_anom,   dtype=int)    ])
```

```python
threshold = 4 
y_pred = (errors > threshold).astype(int)
```

```python
def scan_thresholds(thresholds):
    best = None
    for thr in thresholds:
        y_pred = (errors > thr).astype(int)
        f1  = f1_score(y_true, y_pred)
        rec = recall_score(y_true, y_pred)      # anomaly recall
        fpr = ( (y_pred==1) & (y_true==0) ).sum() / (y_true==0).sum()
        print(f"thr={thr:5.2f}  F1={f1:6.3f}  recall1={rec:6.3f}  FPR={fpr:6.3f}")
        if best is None or f1 > best[1]:
            best = (thr, f1, rec, fpr)
    print("\nBest by F1:", best)
    return best
```

```python
thresholds = np.arange(1, 31, 0.5)
best = scan_thresholds(thresholds)
```

```python
(threshold, none,no,nope) = best
```

```python
class LSTMClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes, num_layers=2, dropout=0.2):
        super(LSTMClassifier, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True, dropout=dropout)
        self.classifier_head = nn.Sequential(
            nn.Linear(hidden_dim, 32),
            nn.BatchNorm1d(32),       
            nn.ReLU(),                 
            nn.Dropout(0.2),           
            nn.Linear(32, num_classes) 
        )

    def forward(self, x):
        out, _ = self.lstm(x)
        last_timestep_out = out[:, -1, :]
        return self.classifier_head(last_timestep_out)
```

```python
input_dim = 5
hidden_dim = 64
num_classes = 4 
```

```python
classifier = LSTMClassifier(input_dim, hidden_dim, num_classes)
classifier.cuda()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(classifier.parameters(), lr=0.001,weight_decay=1e-4)
```

```python
classifier
```

```python
epochs = 100

train_losses, train_accs = [], []
val_losses, val_accs = [], []

for epoch in range(epochs):
    classifier.train()
    train_loss, correct_train, total_train = 0.0, 0, 0
    
    for seqs, labels in train_anomaly_loader:
        seqs, labels = seqs.cuda(), labels.cuda()
        
        optimizer.zero_grad()
        outputs = classifier(seqs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()
        
    epoch_train_loss = train_loss / len(train_anomaly_loader)
    epoch_train_acc = 100 * correct_train / total_train

    train_losses.append(epoch_train_loss)
    train_accs.append(epoch_train_acc)
    
    classifier.eval()
    val_loss, correct_val, total_val = 0.0, 0, 0
    
    with torch.no_grad():
        for seqs, labels in val_anomaly_loader:
            seqs, labels = seqs.cuda(), labels.cuda()
            
            outputs = classifier(seqs)
            loss = criterion(outputs, labels)
            
            val_loss += loss.cpu().numpy()
            _, predicted = torch.max(outputs.data, 1)
            total_val += labels.size(0)
            correct_val += (predicted == labels).sum().item()
            
    epoch_val_loss = val_loss / len(val_anomaly_loader)
    epoch_val_acc = 100 * correct_val / total_val if total_val > 0 else 0
    

    val_losses.append(epoch_val_loss)
    val_accs.append(epoch_val_acc)
    print(f"Epoch [{epoch+1}/{epochs}] | "
          f"Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.2f}% | "
          f"Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.2f}%")
```

```python
plt.plot(range(epochs),train_losses,label='train')
plt.plot(range(epochs),val_losses,label='val')
plt.title('losses')
plt.legend()
```

```python
plt.plot(range(epochs),train_accs,label='train')
plt.plot(range(epochs),val_accs,label='val')
plt.title('accuracy')
plt.legend()
```

```python
torch.save(classifier.state_dict(), 'grid_anomaly\\lstm_classifier.pth')
```

```python
classifier.eval()

all_preds = []
all_true = []

with torch.no_grad():
    for X_batch, y_batch in test_classifier_loader:  
    
        X_batch = X_batch.cuda()
        y_batch = y_batch.cuda()

        outputs = classifier(X_batch)

        _, preds = torch.max(outputs, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_true.extend(y_batch.cpu().numpy())

all_true = np.array(all_true)
all_preds = np.array(all_preds)
```

```python
print(classification_report(all_true, all_preds))
```
